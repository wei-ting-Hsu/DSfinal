```python
import requests
import time

# 从文件中读取地点名称和经纬度信息
def read_coordinates_from_file(file_path):
    coordinates = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.strip().split(',')
            place_name = parts[0]
            latitude, longitude = map(float, parts[1:])
            coordinates[place_name] = (latitude, longitude)
    return coordinates

# 根据地点名称获取对应的经纬度坐标
def get_coordinates_by_location(location_name):
    if location_name in locations:
        return locations[location_name]
    else:
        print("地点名称无效！")
        return None

# 获取天气数据
def get_weather_by_coordinates(api_key, lat, lon):
    url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(url)
    data = response.json()
    return data

# 控制空调
def control_air_conditioner(temperature):
    try:
        if temperature is not None:
            weather_data = get_weather_by_coordinates(api_key, latitude, longitude)
            if temperature > 20:
                # 控制空调开启
                print(weather_data["name"],"溫度 :", round(temperature,1), "°C")
                print("溫度超過20度，開啟冷氣")
            else:
                # 控制空调关闭
                print(weather_data["name"],"溫度 :", round(temperature,1), "°C")
                print("溫度低於20度，關閉冷氣")
        else:
            print("无法获取温度信息")
    except Exception as e:
        print("控制冷氣出錯：", e)

# 读取地点名称和经纬度信息
file_path = '經緯度參考.txt'
locations = read_coordinates_from_file(file_path)

# 示例地点名称
location_name = "東京"
latitude, longitude = get_coordinates_by_location(location_name)

# 替换为你的 API Key
api_key = "b55132e7d0d008aa945b18e4cd2f5a02"

# 主循环
def server():
    while True:
        try:
            # 获取天气数据
            weather_data = get_weather_by_coordinates(api_key, latitude, longitude)
            if weather_data is not None:
                # 获取温度数据
                temperature_celsius = weather_data["main"]["temp"] - 273.15
                # 控制空调
                control_air_conditioner(temperature_celsius)
        except Exception as e:
            print("Error:", e)
        time.sleep(60)  # 每隔 60 秒监测一次温度

# 运行主程序
server()
```
